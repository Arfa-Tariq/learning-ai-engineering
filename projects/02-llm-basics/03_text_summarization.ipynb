# Summarizing
In this notebook, we will summarize text with a focus on specific topics.

````python
!pip install -q groq

from groq import Groq
from google.colab import userdata

GROQ_API_KEY = userdata.get('groq_key')
client = Groq(api_key=GROQ_API_KEY)
````

````python

def get_completion(prompt, model="llama-3.1-8b-instant"):
  messages = [{"role": "user", "content": prompt}]
  response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=1,)
  return response.choices[0].message.content
````

````python
prod_review = """
Got this panda plush toy for my daughter's birthday,
who loves it and takes it everywhere. It's soft and
super cute, and its face has a friendly look. It's
a bit small for what I paid though. I think there
might be other options that are bigger for the
same price. It arrived a day earlier than expected,
so I got to play with it myself before I gave it
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to only generate a short summary of a product
review from an ecommerce site.

Summarize the review below, which is delimited by triple
backticks, in at most 30 words.

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
for i in range(len(reviews)):
    prompt = f"""
    Your task is to generate a short summary of a product
    review from an ecommerce site.

    Summarize the review below, which is provided delimited by triple
    backticks in at most 20 words.

    Review: ```{reviews[i]}```
    """

    response = get_completion(prompt)
    print(i+1, response, "\n")
````

## Exercise 4: Summary for Different Audiences
- Summarize for a doctor vs a patient vs a business executive.

````python
audiences = ["doctor", "patient", "business executive"]

for audience in audiences:
    prompt = f"""
    Your task is to only generate a short summary of an article on the topic "How
    Artificial Intelligence is transforming the healthcare industry" for a {audience}.

    Summarize the article below, which is delimited by triple
    backticks, in exactly 50 words, using language and details appropriate for this audience.
    Output the summary only.

    Article:
    ```{article}```
    """
    response = get_completion(prompt)
    print(f"For {audience.capitalize()}:")
    print(response, "\n")
````
