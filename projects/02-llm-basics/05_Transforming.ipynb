# LLM Text Transformation Techniques

This notebook demonstrates various text transformation capabilities of Large Language Models, including language translation, grammar correction, tone adjustment, and data format conversion. We will explore practical examples for each of these use cases.

````python
!pip install -q groq

from groq import Groq
from google.colab import userdata

GROQ_API_KEY = userdata.get('groq_key')
client = Groq(api_key=GROQ_API_KEY)
````

````python
def get_completion(prompt, model="openai/gpt-oss-120b"):
  messages = [{"role": "user", "content": prompt}]
  response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=1,)
  return response.choices[0].message.content
````

## Translation

LLMs are trained with sources in many languages. This gives the model the ability to do translation.

````python
prompt = f"""
Translate the following English text to Urdu:
```Please confirm the meeting time.```
"""
response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Tell me which language this is:
```안녕하세요. 만나서 반가워요.```
"""
response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Translate the following  text to Urdu and Spanish
and English pirate:
```The weather is beautiful today.```
"""
response = get_completion(prompt)
print(response)
````

### Universal Translator
Imagine you are in charge of IT at a large multinational e-commerce company. Users are messaging you with IT issues in all their native languages. Your staff is from all over the world and speaks only their native languages. You need a universal translator!

````python
user_messages = [
  "J'ai un problème avec mon clavier.", # French: I have a problem with my keyboard.
  "La batería de mi portátil no carga.", # Spanish: My laptop battery is not charging.
  "Il software non si installa correttamente.", # Italian: The software is not installing correctly.
  "Moja drukarka nie działa.", # Polish: My printer is not working.
  "我無法連接到互聯網。" # Traditional Chinese: I cannot connect to the internet.
]
````

````python
for issue in user_messages:
    prompt = f"Tell me what language this is: ```{issue}```"
    lang = get_completion(prompt)
    print(f"Original message ({lang}): {issue}")

    prompt = f"""
    Translate the following  text to English \
    and Korean: ```{issue}```
    """
    response = get_completion(prompt)
    print(response, "\n")
````

## Spellcheck/Grammar check.

Here are some examples of common grammar and spelling problems and the LLM's response.

To signal to the LLM that you want it to proofread your text, you instruct the model to 'proofread' or 'proofread and correct'.

````python
text = [
  "He don't like to eats apples.",  # Incorrect verb agreement
  "The cat sat on the mat it was green.", # Missing punctuation
  "Alot of people think that their going to win.",  # Common homonym errors
  "I would of gone if I had knew.", # Incorrect verb tense
  "This is a example of a sentence with a grammatical error.", # Article usage
  "They're going to loose the game.", # Spelling error
  "Their are many ways to spell tommorow."  # Homonym and spelling
]
for t in text:
    prompt = f"""Proofread and correct the following text
    and rewrite the corrected version. If you don't find
    and errors, just say "No errors found". Don't use
    any punctuation around the text:
    ```{t}```"""
    response = get_completion(prompt)
    print(response)
````
