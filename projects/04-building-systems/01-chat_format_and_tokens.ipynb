## Chat Format and Tokens


```python
!pip install -q groq

from groq import Groq
from google.colab import userdata
import tiktoken

GROQ_API_KEY = userdata.get('groq_key')
client = Groq(api_key=GROQ_API_KEY)
```

```python
def get_completion(prompt, model="openai/gpt-oss-120b", temperature=0):
  messages = [{"role": "user", "content": prompt}]
  response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=temperature,)
  return response.choices[0].message.content
```

```python
response = get_completion("What is the capital of Pakistan?")
response
```

```python
response = get_completion("Take the letters in the word lollipop and reverse them")
response
```

## Helper function (chat format)

```python
def get_completion_from_messages(messages,
                                 model="llama-3.3-70b-versatile",
                                 temperature=0,
                                 max_tokens=8000):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
        max_tokens=max_tokens, # the maximum number of tokens the model can ouptut
    )
    return response.choices[0].message.content
```

```python
# combined
messages =  [
{'role':'system',
 'content':"""You are an assistant who \
responds in the style of Dr Seuss. \
All your responses must be one sentence long."""},
{'role':'user',
 'content':"""write me a story about a happy carrot"""},
]
response = get_completion_from_messages(messages,
                                        temperature =1)
print(response)
```

## Helper function to count the number of tokens


```python
def get_completion_and_token_count(messages,
                                 model="llama-3.3-70b-versatile",
                                 temperature=0,
                                 max_tokens=8000):
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
        max_tokens=max_tokens, # the maximum number of tokens the model can ouptut
    )
    content = response.choices[0].message.content
    token_dict = {
'prompt_tokens':response.usage.prompt_tokens,
'completion_tokens':response.usage.completion_tokens,
'total_tokens':response.usage.total_tokens,
    }

    return content, token_dict
```

```python
messages = [
{'role':'system',
 'content':"""You are an assistant who responds
 in the style of Dr Seuss."""},
{'role':'user',
 'content':"""write me a very short poem
 about a happy carrot"""},
]
response, token_dict = get_completion_and_token_count(messages)
```
